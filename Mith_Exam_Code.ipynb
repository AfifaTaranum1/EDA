# Deleting null max values columns

```python
#drop columns with large number of missing values
df = df.drop(['weight','medical_specialty' ], axis = 1)
```

```python
# Creating a boolean "mask" for each variable
mask_race = df['race'] == '?'
mask_diag1 = df['diag_1'] == '?'
mask_diag2 = df['diag_2'] == '?'
mask_diag3 = df['diag_3'] == '?'
mask_gender = df['gender'] == 'Unknown/Invalid'
```

```python
combined_mask = np.logical_or.reduce([mask_race, mask_diag1, mask_diag2, mask_diag3, mask_gender])
drop_indices = df[combined_mask].index
```

```python
df.drop(drop_indices, inplace=True)
```

```python
df.columns
```

```python
df['race'] = df['race'].fillna(df['race'].mode()[0])
```

# Feature Selection

```python
df = pd.DataFrame(df)

df = df.drop(['index', 'encounter_id', 'patient_id','diag_1', 'diag_2', 'diag_3', 'diag_4', 'diag_5',
       'number_diagnoses', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8',
       'X9', 'X10', 'X11', 'X12', 'X13', 'X14', 'X15', 'X16', 'X17', 'X18',
       'X19', 'X20', 'X21', 'X22', 'X23', 'X24', 'X25'], axis='columns')

print(df)
```

# Recoding variable

```python
df.columns
```

```python
keys = ['race', 'gender', 'age', 'time_in_hospital', 'num_lab_procedures',
       'num_procedures', 'num_medications', 'number_outpatient',
       'number_emergency', 'number_inpatient', 'change', 'diabetesMed',
       'readmitted']
```

```python
df['change'] = df['change'].replace('Ch', 1)
df['change'] = df['change'].replace('No', 0)
df['gender'] = df['gender'].replace('Male', 1)
df['gender'] = df['gender'].replace('Female', 0)
df['diabetesMed'] = df['diabetesMed'].replace('Yes', 1)
df['diabetesMed'] = df['diabetesMed'].replace('No', 0)

for col in keys:
    df[col] = df[col].replace('No', 0)
    df[col] = df[col].replace('Steady', 1)
    df[col] = df[col].replace('Up', 1)
    df[col] = df[col].replace('Down', 1)
```

```python
df['readmitted'] = df['readmitted'].replace({'<30': 1, '>30': 0, 'NO': 0})
df['readmitted'] = df['readmitted'].map({0: 0, 1: 1}).fillna(0)
```

# Training Model

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.preprocessing import StandardScaler 

from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,f1_score
```

```python
X = df.drop('readmitted',axis=1)

y = df['readmitted']
```

```python
X.columns
```

```python
X.dtypes
```

```python
dummies = pd.get_dummies(X,columns = ['race', 'gender', 'age','change', 'diabetesMed'],drop_first = True)
dummies
```

```python
X = pd.concat([X,dummies],axis = 1)
X
```

```python
X.drop(columns = ['race', 'gender', 'age','change', 'diabetesMed'],inplace= True)
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# Model of Logistics Regression

```python
LR = LogisticRegression()

LR.fit(X_train,y_train)
```

```python
pred = LR.predict(X_test)
```

```python
accuracy_score(y_test,pred)
```

```python
print("Classification Report:")
print(classification_report(y_test, pred))
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, pred))
```

```python
print(f"F1 Score: {f1_score(y_test, pred)}")
```

# Model Of Random Forest 

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Predictions on the validation set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print(f"F1 Score: {f1_score(y_test, y_pred)}")
```

```python
accuracy_score(y_test,y_pred)
```
